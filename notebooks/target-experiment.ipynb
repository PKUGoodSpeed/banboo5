```python
def get_price_move_ratio(df, price_col, horizon, sliding_aggr_mode='mean'):
    next_day_price = df[price_col].shift(-1).astype(float)
    future_xday_agg = df[price_col].rolling(horizon).agg(sliding_aggr_mode).shift(-horizon).astype(float)
    ratio = future_xday_agg / next_day_price
    df['next_day_price'] = next_day_price
    df[f'future_{horizon}day_{sliding_aggr_mode}'] = future_xday_agg
    df[f'{sliding_aggr_mode}_ratio'] = ratio
    return df
```

```python
test = get_price_move_ratio(test, 'increase', 3)
test
```
